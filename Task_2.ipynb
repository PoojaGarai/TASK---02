```python
# Identify columns with missing values
missing_values = df.isnull().sum()
print("\nColumns with missing values:")
print(missing_values[missing_values > 0])
```

```python
# Fill missing values with the mean for columns with fewer missing values
columns_to_fill = missing_values[(missing_values < 50) & (df[missing_values.index].dtypes != 'object')].index # Filter out columns with object dtype
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].mean())
```

```python
# Fill columns with many missing values with a placeholder
columns_with_many_missing = missing_values[missing_values >= 50].index
df[columns_with_many_missing] = df[columns_with_many_missing].fillna(-999)
```

```python
print("\nDataFrame after handling missing values:")
print(df.head())
```
